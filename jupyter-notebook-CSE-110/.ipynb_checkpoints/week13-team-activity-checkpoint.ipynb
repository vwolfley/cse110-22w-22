### Core Requirments 1 & 2

```python
#######
# Date: 2022-03-128
# Week 13 Team Activity
# File: week13-team-activity.py
# Author: Vern Wolfley
# Purpose: Areas of Shapes Revisited
# Core Requirements 1 & 2
#######


# Square area of computation
def get_square_area(side):
    sq_area = float(side)**2
    return sq_area

length = input("What is the length of one side of your square? ")
area = get_square_area(length)
print(f'*** The area of your square is: {area} ***')

# Rectangle area of computation
def get_rectangle_area(length, width):
    r_len = float(length)
    r_wid = float(width)
    r_area = round((r_len * r_wid), 3)
    return r_area

r_length = input("What is the length of your rectangle? ")
r_width = input("What is the width of your rectangle? ")
area_of_rectange = get_rectangle_area(r_length, r_width)
print(f'*** The area of your rectangle is: {area_of_rectange } ***')


import math

# Circle area of computation
def get_circle_area(radius):
    cir_area = round((math.pi * float(radius)**2), 3)
    return cir_area

radius = input("What is the radius of your circle? ")
area_of_circle = get_circle_area(radius)
print(f'*** The area of your circle is: {area_of_circle} ***')

```

```python
#######
# Date: 2022-03-128
# Week 13 Team Activity
# File: week13-team-activity.py
# Author: Vern Wolfley
# Purpose: Areas of Shapes Revisited
# Core Requirements 3
#######


# Square area of computation
def get_square_area(side):
    sq_area = float(side)**2
    return sq_area

# Rectangle area of computation
def get_rectangle_area(length, width):
    r_len = float(length)
    r_wid = float(width)
    r_area = round((r_len * r_wid), 3)
    return r_area

import math

# Circle area of computation
def get_circle_area(radius):
    cir_area = round((math.pi * float(radius)**2), 3)
    return cir_area




options = (f'''
{"*":*^80}
What kind of shape do you have:
1. Square
2. Rectangle
3. Circle
4. Quit
{"*":*^80}''')

not_valid = (f''' 
{"*":*^80}
*{"Sorry that is not a valid choice!":^78}*
{"*":*^80}''')

end = (f''' 
{"*":*^80}
*{"Thank you - Goodbye":^78}*
{"*":*^80}''')

action = None
while action != 4:
    while True:
        try:
            print(options)
            action = int(input("\nPlease enter your choice: "))
    
            if action == 1:
                length = input("What is the length of one side of your square? ")
                area = get_square_area(length)
                print(f'\n*** The area of your square is: {area} ***')
            elif action == 2:
                r_length = input("What is the length of your rectangle? ")
                r_width = input("What is the width of your rectangle? ")
                area_of_rectange = get_rectangle_area(r_length, r_width)
                print(f'\n*** The area of your rectangle is: {area_of_rectange } ***')
            elif action == 3:
                radius = input("What is the radius of your circle? ")
                area_of_circle = get_circle_area(radius)
                print(f'*** The area of your circle is: {area_of_circle} ***')
            elif action == 4:
                print(end)

            while action not in [1,2,3,4]:
                print(not_valid)
                break
            break
        except ValueError:
            print(not_valid)
```

### Stretch Callenge 2

```python
#######
# Date: 2022-03-128
# Week 13 Team Activity
# File: week13-team-activity.py
# Author: Vern Wolfley
# Purpose: Areas of Shapes Revisited
# Stretch Challenge 2
#######


def compute_area(shape, length):
    # Square area of computation
    if shape.lower() == "square":
        side = float(length)
        sq_area = side * side
        return sq_area, shape
    # Circle area of computation
    elif shape.lower() == "circle":
        import math
        cir_area = round((math.pi * float(length)**2), 3)
        return cir_area, shape

q1 = input("What shape do you want? (square/circle)  ")
q2 = input("\nWhat is your length?  ")
area, shape = compute_area(q1, q2)

print(f'\n*** The area of your {shape} is: {area} ***')





```

### Stretch Callenge 3

```python
#######
# Date: 2022-03-128
# Week 13 Team Activity
# File: week13-team-activity.py
# Author: Vern Wolfley
# Purpose: Areas of Shapes Revisited
# Stretch Challenge 3
#######

# Compute the area of square, rectangle, circle function
# Takes arguments of shape, length, width
def compute_area(shape, length, width = 0):
    # Square area of computation
    if shape.lower() == "square":
        side = float(length)
        sq_area = side * side
        return sq_area, shape
    # Rectangle area of computation
    elif shape.lower() == "rectangle":
        r_len = float(length)
        r_wid = float(width)
        r_area = round((r_len * r_wid), 3)
        return r_area, shape
    # Circle area of computation
    elif shape.lower() == "circle":
        import math
        cir_area = round((math.pi * float(length)**2), 3)
        return cir_area, shape


options = (f'''
{"*":*^80}
What kind of shape do you have:
1. Square
2. Rectangle
3. Circle
4. Quit
{"*":*^80}''')

not_valid = (f''' 
{"*":*^80}
*{"Sorry that is not a valid choice!":^78}*
{"*":*^80}''')

end = (f''' 
{"*":*^80}
*{"Thank you - Goodbye":^78}*
{"*":*^80}''')

action = None
while action != 4:
    while True:
        try:
            print(options)
            action = int(input("\nPlease enter your choice: "))
    
            if action == 1:
                length = input("What is the length of one side of your square? ")
                area, shape = compute_area("square", length)
            elif action == 2:
                length = input("What is the length of your rectangle? ")
                width = input("What is the width of your rectangle? ")
                area, shape = compute_area("rectangle", length, width)
            elif action == 3:
                radius = input("What is the radius of your circle? ")
                area, shape = compute_area("circle", radius)
            elif action == 4:
                print(end)
                break
                
            print(f'\n*** The area of your {shape} is: {area} ***') 

            while action not in [1,2,3,4]:
                print(not_valid)
                break
            break
        except ValueError:
            print(not_valid)
            

```
