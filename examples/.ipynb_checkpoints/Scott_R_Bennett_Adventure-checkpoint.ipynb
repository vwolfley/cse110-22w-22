```python
yes_ans = ['yes', 'ye', 'y', 'yeah', 'yea', 'yep', 'yup']
no_ans = ['no', 'n', 'na', 'nah', 'nope', 'nop']
cont_1 = False
cont_2 = False
cont_3 = False
end = 'CHASE END.'
else_end = 'You hesitate as a feeling of extreme decision paralysis overcomes you, allowing your pursuer to catch up to you!'
 
print("Welcome to THE CHASE, a Choose Your Own Adventure short story!")
print()
start = input("Ready to play? ").lower()
 
if start in yes_ans:
    print("Excellent, let's begin!")
elif start in no_ans:
    print("Too bad! Buckle up buttercup, this is gonna be a heck of a ride!")
else:
    print("I don't know what you mean by that, but I don't care! Buckle up buttercup, this is gonna be a heck of a ride!")
 
#End Intro, moving on to Scenario 1.
 
print()
 
print("Your pursuer is close, you can glimspe him when you look back! So you need to keep running. Panting, face slick with sweat, you slide around the corner and dash down the hallway. About thirty feet away, you see the exit door--but it has a padlock on it!")
 
print()
print("Hastily, you pat yourself for anything useful that could get you through the door! In your pockets, you find a stick of DYNAMITE, a miniaturized TIME machine, a FIDDLE, and a bucket of PAINT. Or you could KICK down the door.")
 
print()
option_1 = input("What do you do? ").lower()
 
if option_1 == 'dynamite':
    print("You pull out the stick of dynamite! It only takes a second to light the fuse and throw it down the hallway. However, as you throw it, the realization dawns on you that the dynamite is actually a lot bigger than anticipated. And you are in a confined space, running towards it. Uh-oh.\n")
    print("The dynamite explodes with a tremendous BOOM. The door is practically vaporized, but the blast destroys a majority of the building's lower level too. With an ominous groan, the building tilts over and collapses. Guess you won't be winning any races anytime soon.\n")
    print(end)
 
elif option_1 == 'time':
    print("You grab your pocket TARDIS and hurriedly crank the Years dial back, back...back. Right before you hit the door, you activate it. Flames shoot out from behind you, time fluctuates and warps before you, and you step into the distant past! Where there was a concrete jungle--or where will be a concrete jungle--is a thriving primordial forest. Huge trees, lush ferns, the calls of strange creatures in the distance. You can't run into a door if there is no door, after all!\n")
    print("Grinning at your genius plan, you walk forward, head down, staring at the pocket TARDIS as your reset the year. You are so engrossed in your task that you don't notice the T-Rex right beside you. \n")
    print("Correction: you notice the T-Rex when it eats you. Yikes.\n")
    print(end)
 
elif option_1 == 'paint':
    print("Might as well take a page out of the master trickster's book, right? You saw it work on TV all the time! If they could do it, you can too!\n")
    print("You pull out a brush and bucket of paint. A dip of the brush into the bucket, and three quick strokes later, you have a neat hole in the door. Perfect! Just what you need!\n")
    print("As your pursuer draws near, you take a step back, then confidently leap through the hole in the door as fast as you can! Or you would, if there was a hole. Sorry to say, but Looney Tunes is non-indicative of real-life physics.\n")
    print("Instead of diving out the door, you ram into the still very solid door head-first and are instantly knocked out. That's gonna leave a mark.\n")
    print(end)
 
elif option_1 == 'kick':
    print("Whatever! Why bother with a fancy gadget or tactic that will only screw you over when PURE MANPOWER can get the job done better? Bracing yourself, you charge forward and deliver a form-perfect flying kick to the door. Unfortunately, it's still padlocked, and it feels a whole lot sturdier than it looks.\n")
    print("You fall to the floor, where you groan and clutch your leg while rocking back and forth.The door is unperturbed by your pain. Good luck running away now.\n")
    print(end)
 
elif option_1 == 'fiddle':
    print("You quickly (but carefully) whip out your violin and bow and begin to play a jaunty fiddle tune. The music is energetic and adventurous! Unable to help yourself, you transition from running to doing a lively jig as you approach the door. Apparently the padlock feels similarly about the music, as it too begins to bounce back and forth in time with you. In the process, it comes loose!\n")
    print("Seizing the opportunity, you yank the unlocked padlock off the door and push through the door!\n")
    cont_1 = True
 
else:
    print(else_end)
    print()
    print(end)
 
 
# Scenario 1 End.
# Moving on to Scenario 2.
 
 
if cont_1:
    print()
    print("You find yourself on a sidewalk. A quick scan shows an alleyway STRAIGHT across the street, an intersection to your RIGHT, and a large crowd bustling along the sidewalk to the LEFT.\n")
    option_2 = input("Where do you run next? ").lower()
 
    if option_2 == 'back':
        print("Maybe if you actually go backwards, do a fake out, you can throw off your pursuer! Yeah, that's a brilliant idea!\n")
        print("As you whirl around, preparing to duck and weave, you see your pursuer's hand already reaching out for you! He's too close and going too fast--! There's nothing you can do.\n")
        print(end)
 
    elif option_2 == 'up':
        print("You do have rocket boots on, but they've been banged up quite a bit during your high-speed chase. They should be able to hold up for another firing, though, right?\n")
        print("Alright, fine it's worth the risk! Your pursuer is close, and it's EXHAUSTING to run so much. You click the heels of your shoes, hear the click-hum of the rocket boosters activate, and brace yourself for lift-off.\n")
        print("The acceleration kicks like a mule, but you're off! Up, up...up... Oh no. You look down and see them sparking and sputtering, but instead of the rockets failing, they're intensifying. You hurtle straight upwards, faster and faster, with no way to stop. It only takes a few seconds to rise above the city's tallest skyscraper.\n")
        print("As you desperately try to fix your rocket boots, you catch a flash of movement. You barely have a second to process the sight before a low-flying airplane flies right into you, nose-first. Ouch.")
        print(end)
 
    #Hidden/joke options are done, now we have the standard options. They split off from each other for a sub-scenario, then meet up back in scenario 4.
 
    #Scenario 3a is up, "Straight"
    elif option_2 == 'straight':
        print("You dash across the street and all but fly down the alleyway, but your pursuer chases you doggedly. The trash and litter is easy enough to navigate. However, about halfway down the alley, you spot a chain link fence blocking your way. You need to figure out a way to bypass it!\n")
        print("You wildly look around for anything, and you spot a few (hopefully) useful items poking out of the refuse here. You see the handle bars of a Super POGO Stick, the smooth matte chassis of a GRENADE, and a GLITCH in the Matrix that you could take advantage of. Or you could just CLIMB the darn thing.\n")
        option_3a = input("What do you do? ").lower()
 
        if option_3a == 'pogo':
            print("Aha! Using the Super Pogo Stick will make it a breeze to bounce over the fence! You grab it, ready it, and jump--OH MY GOODNESS. It launches you several yards up, and you barely feel it when it lands and bounces you another several yards into the air. The shocks on this thing are amazing, and it's a blast to bounce on! \n")
            print("Honestly, you are having so much fun that you only belatedly realize that each jump only brings you a couple feet closer to the fence, which gives your pursuer plenty of time to catch up to you! The bouncing was fun while it lasted, at least...")
            print(end)
 
        elif option_3a == 'grenade':
            print("A grenade! Perfect! You should be able to do a grenade jump over the fence. The blast will give you extra momentum and height. You scoop up the grenade, pull the pin, toss it at just the right spot in front of the fence. \n")
            print("With expert timing, you calculate each step necessary for the best jumping point. You reach your spot right in front of the primed grenade and jump upwards as high as you can go. And the grenade detonates at just the right moment! \n")
            print("...But it's also a grenade. That just exploded. Right under you. And you don't have video game physics on your side. Oof. \n")
            print(end)
 
        elif option_3a == 'climb':
            print("Screw it, you're going to just climb the darn fence. Anything else is too risky. It is relatively simple to jump, cling to the fence, and start clambering up. The thin metal is uncomfortable as it digs into your fingers, but it's momentary. Until, as last, you reach the top! \n")
            print("As you haul yourself up and over, you feel something snag! Part of your clothes catches on the top of the fence, leaving you hanging upside down! You struggle and strain to free yourself, but it's too late! Your pursuer is able to catch up. What an undignified way to go...")
            print(end)
 
        elif option_3a == 'glitch':
            print("There, the glitch in the Matrix! You stick your hand in and poke around at the code, fiddling and reshuffling things around. If you can do this right, it will save you a lot of time. All you need to do is...\n")
            print("Yes! The world stutters around you--or rather, you stutter in the world, hopping from point to point with little regard for travel or physics. In the few seconds that you are lagging, you successfully manage to blip through the fence to the other side! \n")
            print("Only, your shoes have clipped into the asphalt, and you stumble forward. Unable to control your momentum, you fall and tumble down the staircase of a subway entrance!")
            cont_2 = True
       
        else:
            print(else_end)
            print()
            print(end)
 
    #Scenario 3b is next, "Right"
    elif option_2 == 'right':
        print("You run to the right, right towards a busy intersection! If you're clever and are able to safely navigate the traffic, you could put some distance between yourself and your pursuer. But you need a strategy, some way to ensure you can get across the street without being turned into a hillbilly's next roadkill stew. \n")
        print("You wildly look around for anything that could help. There are several storefronts that you run past, a few of which have outside stands and displays with various items on them. Various useful items. You spot a pair of TAP shoes, a duplication BUTTON, and a comically large U-shaped MAGNET. And you realize that you have FROGGER downloaded on your phone, too. Lots of potential options!\n")
        option_3b = input("What do you do? ").lower()
       
        if option_3b == 'tap':
            print("Perfect, tap shoes! Now you can dazzle the traffic with your slick routine as you easily tap dance across the intersection. With a practiced flick of your feet, you shed your regular shoes and done the tap shoes. They make enourmously satisfying clicks with every step as you stride out onto the intersection...and begin to DANCE.\n")
            print("You are spectacular! You slide, hop, twirl and more, dodging the honking cars that swerve around you. Each move is crisp and precise, and you're still able to add some personal flair to the dance! \n")
            print("You get so caught up in the moment that, as you are doing a coffee grinder a couple yards from the opposite sidewalk, you fail to notice the bus before it runs into you. At least your tap shoes are left there, intact and unharmed. The same can't be said about you, though. Yeesh...\n")
            print(end)
       
        elif option_3b == 'frogger':
            print("That's right, Frogger is all about dodging traffic! That will definitely help! You whip out your phone and pull the game up. Eyes glued to the screen, you can feel yourself enter the zone as you step into the intersection. Forward, pause, right, right, forward, back, forward, left, forward...\n")
            print("A car honks immediately before running you over. So, uh...maybe don't play a video game the next time you try to cross the street? That's probably a bad idea.")
       
        elif option_3b == 'button':
            print("There, a duplication button! Just what you need for maximizing your odds of making it across the intersection. You grab the button and push it. It shifts, warps, and grows into a perfect duplicate of you. Awesome! The two yous high-five before dashing into traffic. \n")
            print("It's real touch and go for a moment, but you each make it through the cars fairly well, agilely dodging when you're not pausing or sprinting. Alas, near the end, the original you gets hit by a car. It gives your duplicate an opening to reach the sidewalk, and your duplicate keeps on running. But YOU you is dead, so...yeah. Bummer.")
            print(end)
 
        elif option_3b == 'magnet':
            print("Sweet, a huge magnet! That ought to help you deal with the mostly-metal cars. Somehow. It even has a switch on it, Push and Pull. Hm. Pulling might be risky, since that would drag you right into traffic, but pushing...yeah. If you could just push the cars out of the way with the magnet, that would make it so much easier! \n")
            print("You flip the selector to Push, aim the super magnet at the nearest car, then activate it with another button press. It hums, and an instant later you feel it work! \n")
            print("Thing is, for every force there is an equal and opposite reaction. Your body weight pushes against the car, shoving it to the side a bit, but in exchange the several hundred pounds of the car's weight pushing back against the magnet--which you are holding. And the car is traveling at speed, which only adds more force against you.\n")
            print("The supersized magnet slams into your sternum and sends you flying backwards so fast that your shoes are knocked off. You soar into an open subway entrance and tumble down the staircase to the level below!")
            cont_2 = True
       
        else:
            print(else_end)
            print()
            print(end)
 
 
    #Scenario 3c is next, "Left"
    elif option_2 == 'left':
        print("A crowd of people throngs the sidewalk, milling around and moving hither and thither. It will be tricky to get through them fast enough to avoid your pursuer, but if you can pull it off, the crowd would create an excellent obstruction in your wake.\n")
        print("You grab an unattended bag as you dash by and rummage around in its contents. There's a shrink ray, which you could use on the CROWD or on YOURSELF, and a pamphlet from a musical that makes you want to SING. You could just try to PUSH through the crowd, too.\n")
        option_3c = input("What do you do? ").lower()
 
        if option_3c == 'push':
            print('"Outta the way!" You brace yourself and start muscling through the crowd, threading through the press of bodies and shoving them to make enough room when necessary. One of the people you shove is a large man who hardly budges an inch, which stops you dead in your tracks.\n')
            print('You realize he has has HUGE muscles, and he looks none too pleased at your antics. "Settle down." He raises a meaty hand, which curls into a fist. Paralyzed with fright, you cannot move as the fist descends, bonking you on the head and driving you a solid four feet into the sidewalk and instantly knocking you unconscious.\n')
            print(end)
       
        elif option_3c == 'crowd':
            print("You grab the shrink ray and aim it at the crowd. You pull the trigger, and a buzzing sound accompanies a bright green beam that instantly shrinks everything it touches by a factor of ten. Soon enough, the entire crowd has been hit by the shrink ray. The now miniature people start shrieking and running around in a blind panic. It's actually pretty funny, since their voices are pitched several octaves higher, and their tiny legs make them much slower.\n")
            print('In any case, you start picking your way through the crowd. You need to be careful not to step on anyone. That would be awkward. But then you hear a shrill voice squeak out, "Hey, you'"'"'re the one who shrunk us!"\n')
            print('Immediately, a hundred tiny eyes fasten to you. "Get em!" The shrunk people yell out a warcry and charge. You stumble and trip as they tackle you. They quickly dogpile you, which is how you discover that tiny fists are actually quite painful. Needless to say, you are rendered unconscious in short order.\n')
            print(end)
       
        elif option_3c == 'yourself':
            print("If you can use the shrink ray on yourself, it should make it a lot easier to navigate the crowd! You aim the barrel at your chest and pull the trigger. There's a flash of green light and a buzzing noise, and when you blink, you realize that you are about six inches tall. Groovy!\n")
            print("You run into the crowd, but it's slow going. Your legs are tiny now, and the forest of legs surrounding you is pretty intimidating. Suddenly, a foot comes out of nowhere and kicks you into the air! You soar for only a moment, then bounce off another person's leg, which also sends you flying.\n")
            print("This whole sequence repeats like a game of demented pinball, until you are so thoroughly battered that by the time you come to a solid stop, it takes a long moment to realize that you are sitting in a toy box.\n")
            print('The head of a boy enters your vision, and he peers excitedly at you. "Mom, mom! Look at this action figure! It looks so realistic, and the articulation looks amazing! Can I get it? Pleeeaaassse?"')
            print(end)
       
        elif option_3c == 'sing':
            print("Yeah. This is happening.\n")
            print("You stomp your foot and clap your hands overhead, then belt out the opening words to the best song EVER. The crowd comes to an immediate stop. As one, they spring into action, breaking out into song as well, and begin a perfectly choreographed dance along the sidewalk!\n")
            print("You stride forward, and the crowd parts and dances around you. Backup singers support and harmonize with your tune, and the extras whirl and leap behind you. That will without a doubt slow down your pursuer--but you're having way too much fun right now. This is AWESOME.\n")
            print("The sidewalk is awash with movement and song, and you're the one leading it. Its an exhilarating feeling! You could go forever, and you don't want to stop!\n")
            print("Alas, your foot catches in a pothole. You stumble forward uncontrollably, shoe ripped off your foot, and fall down the stairs of a subway entrance!")
            cont_2 = True
       
        # else:
        #     print(else_end)
        #     print()
        #     print(end)
   
    else:
        print(else_end)
        print()
        print(end)
 
 
 
if cont_2:
    print("TEST")
 
 
 
 
```
