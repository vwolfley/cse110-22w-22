### Stretch Challange 03

```python
#######
# Date: 2022-03-15
# Week 12 Team Activity
# File: week12-team-activity.py
# Author: Vern Wolfley
# Purpose: Scripture Books
# Stretch Requirements
#######

# Path to my directory
file_path = "../assignment_files/books_and_chapters.txt"
book_files = file_path

scripture = None
chapter = 0
book = None

scripture_max = None
chapter_max = 0
book_max = None

unique_scriptures = []
scripture_list = []
user_input = None

while user_input != "Quit":
    
    with open(book_files) as books:
        for item in books:
            item = item.strip()
            item = item.split(":")
#             print(item)

            # Create a unique list of scriptures to guide user in choice
            unique_scriptures.append(item[2].title())

    scripture_list = list(set(unique_scriptures))
#     print(scripture_list)

    for i in scripture_list:
        print(i)
    
    user_input = input(f"\nWhat volume of scripture do you want to look at?  ").title()
    if user_input not in scripture_list and user_input != "Quit":
        print("\n*** Sorry not a valid scripture, please try again! ***\n")
    elif user_input == "Quit":
        print("\n*** Thank You! Have a Nice Day! ***")
    else:
        with open(book_files) as books:
            for item in books:
                item = item.strip()
                item = item.split(":")
#                 print(item)

                if item[2].title() == user_input:
                    scripture = item[2]
                    chapter = int(item[1])
                    book = item[0]
                    
                    # *** Stretch Requirement 01 ***
#                     print(f"Scripture: {scripture:<20} Book: {book:<15} Chapter: {chapter}")
                   
                    # Find the max
                    if chapter > chapter_max:
                        scripture_max = scripture
                        chapter_max = chapter
                        book_max = book

        # *** Stretch Requirement 02 ***
        max = f'''\n*** The {scripture_max}, book of {book_max}, has the largest number of chapters at {chapter_max}. ***\n'''
        print(max)




```

### Instructor Example

```python
"""
File: teach12_sample.py
Author: Brother Burton

Purpose: Practice finding items in files.
"""
# Path to my directory
file_path = "../assignment_files/books_and_chapters.txt"
book_files = file_path


chosen_volume = input("Which volume of scripture would you like to learn about? ")

max_chapters = -1
book_with_max = ""

# Open the file with the information
with open(book_files) as books_file:

    # Iterate through the file one line at a time
    for line in books_file:
        # Split up the line based on the ":"
        parts = line.split(":")

        # Get the book and strip off any leading/trailing whitespace
        book = parts[0].strip()

        # Get the number of chapters as an integer
        chapters = int(parts[1])

        # Get the volume of scripture and strip off any leading/trailing whitespace
        scripture = parts[2].strip()

        # Check if this belongs to the user's chosen volume:
        if scripture.lower() == chosen_volume.lower():
            # Only display and check for the max inside the body
            # of this if statement. That way we are limiting it
            # to the user's chosen volume.

            print(f"Scripture: {scripture}, Book: {book}, Chapters: {chapters}")

            # Check to see if this book has the most chapters we've seen so far
            if chapters > max_chapters:
                # This book is the new best one!
                max_chapters = chapters # The max chapters is now this one
                book_with_max = book # Remember the name of the book

# This is now after the loop has finished
print(f"The book with the most chapters in the {chosen_volume} is: {book_with_max}")
print(f"It has {max_chapters} chapters.")
```
