### Assignment

Ask the user for a list of list of items in a shopping list, stop asking for items when the user types "quit."

Then complete the following:

01. Loop through the items in the regular way (for example, for thing in the_list) and display each one to make sure you have the initial list built correctly.

02. Add another loop to go through and print the items in the list, but this time, instead of using the for ... in syntax, use an index (for example, for ... in range) and then access the item using the index and the square brackets. Print the index in front of the items like so: 0. Milk

03. Ask the user for an index and a new shopping list item. Replace the item at that index with the new item. Then, display the whole list again.

The following examples show the expected output:
```
Please enter the items of the shopping list (type: quit to finish):
Item: Milk
Item: Bread
Item: Candy
Item: Carrots
Item: quit

The shopping list is:
Milk
Bread
Candy
Carrots

The shopping list with indexes is:
0. Milk
1. Bread
2. Candy
3. Carrots

Which item would you like to change? 2
What is the new item? Apples

The shopping list with indexes is:
0. Milk
1. Bread
2. Apples
3. Carrots
```


````python
"""
File: week10-prepare
Author: Vern Wolfley
Purpose: Practicing with list indexes.
"""

shopping_list = []

item = None
while item != "Quit":
    item = input("What item to you want in your cart? (Type 'Quit' to finish)  ").capitalize()
    if item != "Quit":
        shopping_list.append(item)

print("\nYour shopping list:")
for item in shopping_list:
    print("Item - ", item )

print("\nYour shopping list with indexes:")
for i in range(len(shopping_list)):
    item = shopping_list[i]
    print(f"{i}. {item}")
    
change = int(input(f"\nWhich item would you like to change? (0-{len(shopping_list)-1})) "))
shopping_list.pop(change)

add_item = input("\nWhat item would you like to add?  ").capitalize()
shopping_list[change].add_item

print("\nYour new shopping list with indexes:")
for i in range(len(shopping_list)):
    item = shopping_list[i]
    print(f"{i}. {item}")
````
